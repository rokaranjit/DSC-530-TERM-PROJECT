```python
# Selected variables
variables = ['Charging Rate (kW)', 'Energy Consumed (kWh)', 'Charging Duration (hours)', 'Battery Capacity (kWh)', 'Temperature (°C)']

# Calculate mean, mode, spread (standard deviation), and tails (5th and 95th percentiles) for each variable
for var in variables:
    print(f"Descriptive statistics for {var}:")
    
    # Mean
    mean_val = df[var].mean()
    print(f"Mean: {mean_val}")
    
    # Mode
    mode_val = df[var].mode().iloc[0] if not df[var].mode().empty else "No mode"
    print(f"Mode: {mode_val}")
    
    # Spread (Standard Deviation)
    std_dev = df[var].std()
    print(f"Spread (Standard Deviation): {std_dev}")
    
    # Tails (5th and 95th percentiles)
    lower_tail = df[var].quantile(0.05)
    upper_tail = df[var].quantile(0.95)
    print(f"5th Percentile: {lower_tail}")
    print(f"95th Percentile: {upper_tail}")
    
    print("\n")
```

```python
# Define function to identify outliers using IQR
def find_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers in '{column}':")
    print(outliers[column])
    print(f"Number of outliers in '{column}': {len(outliers)}")
    print("\n")

# List of variables to check for outliers
variables = ['Charging Rate (kW)', 'Energy Consumed (kWh)', 'Charging Duration (hours)', 'Temperature (°C)']

# Identify outliers for each variable
for var in variables:
    find_outliers(df, var)
```

# Normal or Gaussian Distribution

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Extract the 'Temperature (°C)' variable from the dataset
temperature_data = df['Temperature (°C)']

# Fit the data to a normal distribution and get the mean and standard deviation
mu, std = stats.norm.fit(temperature_data)

# Generate the x values for plotting the normal distribution curve
xmin, xmax = min(temperature_data), max(temperature_data)
x = np.linspace(xmin, xmax, 100)
pdf_fitted = stats.norm.pdf(x, mu, std)

# Plot the histogram and the normal distribution curve
plt.figure(figsize=(10, 6))
plt.hist(temperature_data, bins=20, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.plot(x, pdf_fitted, 'r--', label=f'Normal Fit: μ={mu:.2f}, σ={std:.2f}')
plt.title('Temperature (°C) Distribution Fit to Normal Curve')
plt.xlabel('Temperature (°C)')
plt.ylabel('Density')
plt.legend()
plt.show()

# Perform a goodness-of-fit test (Kolmogorov-Smirnov test)
ks_stat, p_value = stats.kstest(temperature_data, 'norm', args=(mu, std))

# Print the result of the goodness-of-fit test
print(f"Kolmogorov-Smirnov Test Statistic: {ks_stat}")
print(f"P-value: {p_value}")

if p_value > 0.05:
    print("The data follows a normal distribution (fail to reject H0).")
else:
    print("The data does not follow a normal distribution (reject H0).")
```

This chart displays a histogram of the Temperature (°C) variable overlaid with a fitted normal distribution curve (dashed red line). Here’s an interpretation of the findings:

Shape Comparison: The histogram bars do not closely follow the normal curve, indicating that the Temperature (°C) data does not perfectly match a normal distribution. The actual data is more uniformly spread across a range, especially at lower temperature values, rather than peaking near the mean.

Mean and Standard Deviation: The red dashed line represents a normal distribution fit with a mean (μ) of 15.26 and a standard deviation (σ) of 14.83. These values describe the center and spread of the fitted normal distribution but do not capture the actual distribution well.

Skewness: The histogram shows a slightly left-skewed distribution with a more gradual decline as temperatures increase, suggesting a departure from normality. There are some higher temperatures that are less frequent, contributing to the longer right tail.

Implications for Analysis: Since the Temperature (°C) data does not align well with a normal distribution, using normal-based statistical methods might not be ideal. Further analysis could explore other distributions that might better fit this data.

# Plotting Cumulative Distribution Function (CDF) & Probability Mass Function (PMF)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Ensure Charging Start Time and Charging End Time are in datetime format
df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'])
df['Charging End Time'] = pd.to_datetime(df['Charging End Time'])

# Calculate Charging Duration in hours
df['Charging Duration (hours)'] = (df['Charging End Time'] - df['Charging Start Time']).dt.total_seconds() / 3600

# Create PMF
pmf_values, bins = np.histogram(df['Charging Duration (hours)'], bins=30, density=True)

# Calculate the width of each bin for PMF calculation
bin_width = np.diff(bins)

# Create CDF
cdf_values = np.cumsum(pmf_values * bin_width)  # Multiply PMF by bin width for proper scaling

# Plot PMF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(bins[:-1], pmf_values, width=bin_width, edgecolor='black', alpha=0.7, align='edge')
plt.xlabel('Charging Duration (hours)')
plt.ylabel('PMF')
plt.title('Probability Mass Function of Charging Duration')

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(bins[:-1], cdf_values, marker='o', color='blue', linestyle='-')
plt.xlabel('Charging Duration (hours)')
plt.ylabel('CDF')
plt.title('Cumulative Distribution Function of Charging Duration')

plt.tight_layout()
plt.show()
```

# Hypothesis Testing

```python
from scipy.stats import ttest_ind

# Hypothesis test: Comparing Charging Rate for high vs low temperatures
t_stat, p_val = ttest_ind(high_temp, low_temp)
print("T-Statistic:", t_stat)
print("P-Value:", p_val)
```

Interpretation:
P-Value (0.835): The p-value is much higher than the common significance levels (e.g., 0.05 or 0.01), which means we fail to reject the null hypothesis. In other words, there's no statistically significant difference between the charging rates for high and low temperatures. This suggests that, based on this data, temperature might not have a significant impact on charging rate.

T-Statistic (-0.208): The t-statistic here is very close to zero, reinforcing the result that there’s little difference between the means of the two groups (high vs. low temperatures).

Conclusion:
Since the p-value is large, the results suggest that any observed difference in charging rate between high and low temperatures is likely due to chance rather than a true effect of temperature. Temperature, based on this analysis, does not appear to significantly influence the charging rate.
